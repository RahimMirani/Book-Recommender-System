Popularity based recommendation system  

```python
BookRating = Ratings.merge(Books, on='ISBN')
total_ratings = BookRating.groupby('Book-Title').count()['Book-Rating'].reset_index()
total_ratings.rename(columns={'Book-Rating':'Rating_count'}, inplace=True)
```

```python
# Convert the Book-Rating column to numeric, invalid values will be NaN
BookRating['Book-Rating'] = pd.to_numeric(BookRating['Book-Rating'], errors='coerce')

#Dropped NaN values
BookRating_clean = BookRating.dropna(subset=['Book-Rating'])

# Now calculate the average ratings
avg_ratings = BookRating_clean.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
avg_ratings.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)
avg_ratings
```

```python
#Merging total number of rating and average rating df
popularity_df = total_ratings.merge(avg_ratings,on= 'Book-Title')
popularity_df
```

```python
popularity_df=popularity_df[popularity_df['Rating_count'] >= 250].sort_values('avg_rating', ascending=False).head(50)
popularity_df=popularity_df.merge(Books, on = 'Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df=popularity_df[['Book-Title', 'Book-Author', 'Image-URL-M', 'Rating_count', 'Rating_count']]
popularity_df
```
